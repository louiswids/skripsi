```python
import pandas as pd
import seaborn as sns
```

Article was acquired from: https://pubmed.ncbi.nlm.nih.gov/18338325/

```python
data = {
    "Batch": [1, 2, 3],
    "0": [101, 102, 100],
    "3": [97, 100, 104],
    "6": [96, 99, 96],
    "9": [95, 100, 98],
    "12": [95, 102, 97],
    "18": [94, 102, 96],
    "24": [92, 100, 99],
    "36": [92, 101, 97]
}


df = pd.DataFrame(data)


df = df.melt(id_vars=["Batch"], 
             var_name="Time", 
             value_name="Potency Assay")


df["Time"] = df["Time"].astype(int)

df
```

```python
sns.lineplot(data = df, x = 'Time',y = 'Potency Assay', hue= 'Batch',marker = 'o')
```

Potency Assay for this literature is very fluctuating
